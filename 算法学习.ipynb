# 智能刷题系统 - 算法学习

这个 notebook 演示如何使用智能刷题系统来练习算法题目。

````python
import sys
import os

# 添加当前目录到路径
sys.path.append(os.path.dirname(os.path.abspath('__file__')))

from question_importer import QuestionBankImporter
from practice_session import PracticeSession, QuestionPractice
from database import DatabaseManager

print("✅ 智能刷题系统模块加载成功！")
````

````python
# 示例：导入一些算法题目
sample_questions = """
1. 以下哪种排序算法的平均时间复杂度是O(nlogn)？
A. 冒泡排序
B. 快速排序
C. 选择排序
D. 插入排序
答案：B

2. 二叉树的前序遍历顺序是？
A. 左-根-右
B. 根-左-右
C. 左-右-根
D. 右-根-左
答案：B

3. 以下哪个数据结构适用于实现LRU缓存？
A. 栈
B. 队列
C. 哈希表+双向链表
D. 二叉搜索树
答案：C
"""

print("示例题目已准备，包含3道算法基础题")
print("\n题目预览：")
print("1. 排序算法时间复杂度")
print("2. 二叉树遍历")
print("3. LRU缓存实现")
````

````python
# 导入题目到数据库
importer = QuestionBankImporter()
count, errors = importer.import_from_text(sample_questions)
importer.close()

print(f"✅ 成功导入 {count} 道题目")
if errors:
    print(f"⚠️  {len(errors)} 道题目导入失败")
````

````python
# 查看题库统计
db = DatabaseManager()
stats = db.get_statistics()
db.close()

print("📊 题库统计：")
print(f"  总题目数: {stats.get('total_questions', 0)}")
print(f"  已练习: {stats.get('practiced', 0)}")
print(f"  正确率: {stats.get('accuracy', 0):.1%}")
````

````python
# 获取一道随机题目进行练习
session = PracticeSession()
questions = session.get_questions('random', 1)
session.close()

if questions:
    q = questions[0]
    print("📝 练习题目：")
    print(f"\n{q['question_content']}\n")
    
    # 显示选项
    import json
    opts = q.get('options', {})
    if isinstance(opts, str):
        opts = json.loads(opts)
    
    if isinstance(opts, dict):
        for k, v in opts.items():
            print(f"  {k}. {v}")
    
    print(f"\n💡 正确答案: {q['correct_answer']}")
else:
    print("题库为空，请先导入题目")
````

## 🎯 使用说明

1. **导入题目**: 使用 `QuestionBankImporter` 导入文本格式的题目
2. **开始练习**: 使用 `PracticeSession` 获取题目
3. **检查答案**: 使用 `QuestionPractice` 验证答案
4. **Web界面**: 运行 `streamlit run web_interface.py` 启动完整界面
